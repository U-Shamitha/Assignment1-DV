```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# **Question 1**  
Plot a line plot between a and b:

a=np.arange(40,50)   
b=np.arange(50,60)


```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.plot(a,b,'r*--', linewidth=1, markersize=5)
plt.title('Line Plot')
plt.xlabel('a')
plt.ylabel('b')
```

# **Question 2**  
Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week

sales_1 = [160,150,140,145,175,165,180] #sales of company1   
sales_2 = [70,90,160,150,140,145,175] #sales of company2


```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2

```

```python
plt.plot(days,sales_1,'g.-.', linewidth=1, markersize=5,label="sales of company1")
plt.plot(days,sales_2,'r*--', linewidth=1, markersize=5,label="salesof company2")
plt.title('Sales of company1&2')
plt.xlabel('Days')
plt.ylabel('Sales')
plt.legend()
```

# **Question 3**  
Create a 3 by 3 subplots:
multiple plots

x = [1,2,3,4]

y2 = [10,20,30,40]  
y3 = [40,30,20,10]

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

```python
## Creating Subplots
plt.figure(figsize=(9,7))
plt.subplot(3,3,1)
plt.plot(x,y1,'r-.')
plt.subplot(3,3,3)
plt.plot(x,y2,'g*--')
plt.subplot(3,3,5)
plt.plot(x,y3,'bo')
plt.subplot(3,3,7)
plt.plot(x,y4,'black',marker='s',linestyle=':')
plt.subplot(3,3,9)
plt.plot(x,y5,color='grey',marker='^')
```
